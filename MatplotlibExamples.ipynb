### Stacked Bar Graph

```python
# color and label are optional parameters
x1=[1,2,3,4,5]
y1=[5,8,2,4,6]
x2=[1,2,3,4,5]
y2=[3,6,4,2,9]
plt.bar(x1,y1,label="Bars1", color='r')
plt.bar(x2,y2,label="Bars2", color='#225533')
plt.legend()
```

### Stacked Area Plot

```python
days = [1,2,3,4,5]
sleeping = [3,5,8,9,4]
eating = [2,3,4,3,2]
working = [6,8,6,2,2]
playing =[8,5,6,8,11]
plt.plot([],[], color='r', label='Sleeping', linewidth = 5)
plt.plot([],[], color='g', label='Eating', linewidth = 5)
plt.plot([],[], color='orange', label='Working', linewidth = 5)
plt.plot([],[], color='b', label='Playing', linewidth = 5)
plt.legend()
plt.stackplot(days, sleeping, eating, working, playing, colors=['r','g','orange','c'])
```

### Pie Chart

```python
days = [1,2,3,4,5]
sleeping = [3,5,8,9,4]
eating = [2,3,4,3,2]
working = [6,8,6,2,2]
playing =[8,5,6,8,11]

slices = [6,2,2,11]
activities = ['sleeping', 'eating', 'working', 'playing']
clrs=['r','g','orange','c']
plt.pie(slices, labels=activities, startangle=90, colors=clrs, shadow=True, explode=(0,0,0.1,0), autopct='%1.1f%%')
```

### Subplots

```python
import random 
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure();

def create_plot(): 
    xs = []
    ys = []
    
    for i in range(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
    return xs, ys

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)
#ax4 = fig.add_subplot(224)

x,y = create_plot()
ax1.plot(x,y)

x,y = create_plot()
ax2.plot(x,y)

x,y = create_plot()
ax3.plot(x,y) 

#x,y = create_plot()
#ax4.plot(x,y) 
```

```python
# subplot2grid()

style.use('fivethirtyeight')

fig = plt.figure();

def create_plot(): 
    xs = []
    ys = []
    
    for i in range(10):
        x = i
        y = random.randrange(10)
        
        xs.append(x)
        ys.append(y)
    return xs, ys

ax1 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((6,1), (1,0), rowspan=4, colspan=1)
ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)

x,y = create_plot()
ax1.plot(x,y)

x,y = create_plot()
ax2.plot(x,y)

x,y = create_plot()
ax3.plot(x,y) 
```
