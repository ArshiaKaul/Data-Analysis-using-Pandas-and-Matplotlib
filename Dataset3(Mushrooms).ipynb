# Edible and Poisonous mushrooms based on cap-color

#### The number of edible and poisonous samples corresponding to each cap color is to be found out and represented graphically
The following two steps need to be done separately for edible and poisonous mushrooms
* Extract the list of number of samples for each color and sort it.
* Extract the list of colors in the same order as their corresponding values.
* Create list of proper color names in the same order as their codes appear in the extracted list. This list will be used in the graph as tick label.


```python
# pandas Series of number of samples for each cap color in sorted order
edible_color_value_series = maindf.loc[maindf["class"]=='e'].groupby(by=["class", "cap-color"])["cap-color"].count().sort_values()
# list of number of samples for each cap color in sorted order
edible_color_value_list = maindf.loc[maindf["class"]=='e'].groupby(by=["class", "cap-color"])["cap-color"].count().sort_values().tolist()
# list of colors in the same order
edible_color = edible_color_value_series.index.get_level_values("cap-color").tolist()
print(edible_color_value_list)
print(edible_color)

# same procedure to obtain value and color list for poisonous mushrooms
poisonous_color_value_series = maindf.loc[maindf["class"]=='p'].groupby(by=["class", "cap-color"])["cap-color"].count().sort_values()
poisonous_color_value_list = maindf.loc[maindf["class"]=='p'].groupby(by=["class", "cap-color"])["cap-color"].count().sort_values().tolist()
poisonous_color = poisonous_color_value_series.index.get_level_values("cap-color").tolist()
print(poisonous_color_value_list)
print(poisonous_color)
```

```python
# create list of color names in the same order as their codes appear in the lists in above cell
edible_color_names=["green", "purple", "cinnamon", "buff", "pink", "yellow", "red", "white", "gray", "brown"]
poisonous_color_names=["cinnamon", "pink", "buff", "white", "yellow", "gray", "red","brown"]

# create a ColumnDataSource and pass the two lists as two columns of the data source
# this data source will be used as a parameter to the graph
edible_source = ColumnDataSource(
    data=dict(
        edible_color_names= edible_color_names,
        edible_color_value_list= edible_color_value_list
    )
)

# create a HoverTool to display the cap color and the corresponding number of samples
# tooltips are created as key-value pairs
# write the column name after '@' 
ehover = HoverTool(tooltips=[
    ("cap color", "@edible_color_names"),
    ("number of samples", "@edible_color_value_list")
])

# create a figure
# specify that the y-axis is categorical by passing the value of y_range as 'edible_color_names'
# pass the ehover tool in the tools list to enable hovering
p = figure(y_range=edible_color_names, plot_width=450, plot_height=400, title="Samples of edible mushrooms based on cap color",
          x_axis_label="number of samples", y_axis_label="cap color", tools=[ehover])

# finally, create horizontal bars
p.hbar(y='edible_color_names', right='edible_color_value_list', height=0.5, color="#2266ff", alpha=0.6, source=edible_source)

# turn the horizontal gridlines invisible as they are not needed
p.ygrid.grid_line_color = None

# add space between ticks and axis label names
p.xaxis.axis_label_standoff = 15
p.yaxis.axis_label_standoff = 20


# go around the same way for the poisonous mushrooms based on cap color
poisonous_source = ColumnDataSource(
    data=dict(
        poisonous_color_names= poisonous_color_names,
        poisonous_color_value_list= poisonous_color_value_list
    )
)

phover = HoverTool(tooltips=[
    ("cap color", "@poisonous_color_names"),
    ("number of samples", "@poisonous_color_value_list")
])

q = figure(y_range=poisonous_color_names, plot_width=450, plot_height=400, title="Samples of poisonous mushrooms based on cap color",
          x_axis_label="number of samples", y_axis_label="cap color", tools=[phover])

q.hbar(y='poisonous_color_names', right='poisonous_color_value_list', height=0.5, color="red", alpha=0.6, source=poisonous_source)

q.ygrid.grid_line_color = None
q.xaxis.axis_label_standoff = 15
q.yaxis.axis_label_standoff = 20

# to show both the plots side by side, use 'gridplot' inside of show()
# eg: show(gridplot( [[p,q],[r,s]] ))
# the first list shows that figure 'p' and 'q' will render in the first row 
# similarly, second list shows that figure 'r' and 's' will render in the second row
# this way, a grid of rows containing multiple plots can be created
show(gridplot([[p,q]]))
```

**Conclusion:** *Slightly higher number of brown and gray mushroom species are edible as compared to their poisonous counterparts. All purple and green colored species analyzed turned out to be edible.*

# Relation between cap color and number of rings

Extract list of colors in the sorted order of the number of records they appear in.

```python
# pandas Dataframe in the order of color code with least number of records to most number of records
maindf.groupby(by=["cap-color"]).count().sort_values(by=["class"])
```

```python
# list of colors in sorted order
# color with code 'r' appears least frequently and 'n' appears most frequently
cap_colors = maindf.groupby(by=["cap-color"]).count().sort_values(by=["class"], ascending=False).index.get_level_values("cap-color").tolist()
cap_colors
```

```python
# grouping mushrooms based on cap color where number of rings is zero
# the Dataframe implies that color codes 'c', 'e', 'n' have the tweleve such records each where number of rings is zero
maindf.loc[maindf["ring-number"]=='n'].groupby(by=["cap-color", "ring-number"]).count()
```

```python
# list of values for the Dataframe above
maindf.loc[maindf["ring-number"]=='n'].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist()
```

```python
# number of records for various colors where number of rings is one
maindf.loc[maindf["ring-number"]=='o'].groupby(by=["cap-color", "ring-number"]).count()
```

```python
# list of values for the Dataframe above
maindf.loc[maindf["ring-number"]=='o'].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist()
```

```python
# number of records for various colors where number of rings is two
maindf.loc[maindf["ring-number"]=='t'].groupby(by=["cap-color", "ring-number"]).count()
```

```python
# list of values for the Dataframe above
maindf.loc[maindf["ring-number"]=='t'].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist()
```

```python
# create three lists
# no_rings stores number of records for each color where number of rings is zero; if no such record is present for a particular 
# color, 0 is appended for that color
# similarly, lists are made for one ring and two rings
no_ring = []
one_ring = []
two_rings = []

# loop over the list of cap colors
# if record/records exist for that cap color and ring number, list with a single value is returned in the 'if' condition, otherwise
# an empty list is returned. In case an empty list is returned, 0 is appended 
for color in cap_colors:
    if(len(maindf.loc[(maindf["ring-number"]=='n') & (maindf["cap-color"]==color)].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist())==0):
        no_ring.append(0)
    else:
        no_ring.append(maindf.loc[(maindf["ring-number"]=='n') & (maindf["cap-color"]==color)].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist()[0])
    
    if(len(maindf.loc[(maindf["ring-number"]=='o') & (maindf["cap-color"]==color)].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist())==0):
        one_ring.append(0)
    else:
        one_ring.append(maindf.loc[(maindf["ring-number"]=='o') & (maindf["cap-color"]==color)].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist()[0])
    
    if(len(maindf.loc[(maindf["ring-number"]=='t') & (maindf["cap-color"]==color)].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist())==0):
        two_rings.append(0)
    else:
        two_rings.append(maindf.loc[(maindf["ring-number"]=='t') & (maindf["cap-color"]==color)].groupby(by=["cap-color", "ring-number"]).count()["class"].tolist()[0])
```

```python
# all lists are of the same length (which is how it should be)
# each list has a value correspondng to all colors in the dataframe
print(len(no_ring))
print(len(one_ring))
print(len(two_rings))
print(len(cap_colors))
```

```python
from bokeh.models import ColumnDataSource, Range1d, LabelSet, Label
from bokeh.transform import dodge
from bokeh.core.properties import value

color_names = ["brown", "gray", "red", "yellow", "white", "buff", "pink", "cinnamon", "purple", "green"]

# create a data dictionory
data = {'color_names' : color_names,
        'none'   : no_ring,
        'one'   : one_ring,
        'two'   : two_rings}

# create a ColumnDataSource
source = ColumnDataSource(data=data)

# LabelSet is used to create labels (in our case, it shows number of samples) 
# these figures are displayed just on top of the bars in the graph below

none_labels = LabelSet(x='color_names', y='none', text='none', level='glyph',
        x_offset=-30, y_offset=0, source=source, render_mode='canvas', text_font_size='8pt')

one_labels = LabelSet(x='color_names', y='one', text='one', level='glyph',
        x_offset=-10, y_offset=0, source=source, render_mode='canvas', text_font_size='8pt')

two_labels = LabelSet(x='color_names', y='two', text='two', level='glyph',
        x_offset=12, y_offset=0, source=source, render_mode='canvas', text_font_size='8pt')

# create a figure
p = figure(x_axis_label="cap color", y_axis_label="number of samples", x_range=color_names, plot_height=400, plot_width=950, 
           title="Cap Color and Number Of Rings", toolbar_location=None)

# create a bar for column in the ColumnDataSource: none, one, two
# each color will have three bars to its name
p.vbar(x=dodge('color_names', -0.25, range=p.x_range), top='none', width=0.2, source=source,
       color="#2255dd", legend=value("no ring"), alpha=0.8)

p.vbar(x=dodge('color_names',  0,  range=p.x_range), top='one', width=0.2, source=source,
       color="#418dff", legend=value("one"), alpha=0.6)

p.vbar(x=dodge('color_names',  0.25, range=p.x_range), top='two', width=0.2, source=source,
       color="#e84d60", legend=value("two"), alpha=0.8)

# padding from the vertical axes 
p.x_range.range_padding = 0.05

# removes vertical gridlines
p.xgrid.grid_line_color = None

# add spacing between ticks and axis label names
p.xaxis.axis_label_standoff = 15
p.yaxis.axis_label_standoff = 20

# change orientation and location of the legend
p.legend.location = "top_right"
p.legend.orientation = "horizontal"

# render the labels
p.add_layout(none_labels)
p.add_layout(one_labels)
p.add_layout(two_labels)

show(p)
```

**Conclusion:** *Most of the samples obtained had a brown cap color. Except for pink capped mushrooms, among rest of the cap colored mushrooms, most of the species had no rings. As for purple and green capped mushrooms, all the species analyzed had no rings.*
